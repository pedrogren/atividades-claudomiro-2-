```python
import random

class SimuladorCache:
    def __init__(self):
        self.ram = {format(i, '05b'): f"DADO_{i}" for i in range(32)}

        self.cache = {
            "00": {"tag": None, "dados": [None, None]},
            "01": {"tag": None, "dados": [None, None]},
            "10": {"tag": None, "dados": [None, None]},
            "11": {"tag": None, "dados": [None, None]}
        }

    def acessar(self, endereco_bin):
        if len(endereco_bin) != 5:
            return "O endereço deve ter 5 bits."

        tag = endereco_bin[0:2]
        linha = endereco_bin[2:4]
        byte = int(endereco_bin[4])

        print(f"Decodificação -> Tag: {tag}, Linha: {linha}, Byte: {byte}")

        linha_cache = self.cache[linha]

        # Comparador
        if linha_cache["tag"] == tag:
            print("ACERTO!")
            dado = linha_cache["dados"][byte]
            return f"Resultado: {dado} (obtido da Cache)"
        else:
            print("MISS!")
            print(f"Buscando bloco na RAM e atualizando linha {linha}...")

            bloco_addr_0 = tag + linha + "0"
            bloco_addr_1 = tag + linha + "1"

            linha_cache["tag"] = tag
            linha_cache["dados"] = [self.ram[bloco_addr_0], self.ram[bloco_addr_1]]

            dado = linha_cache["dados"][byte]
            return f"Resultado: {dado} (obtido da RAM)"

def menu():
    sim = SimuladorCache()
    while True:
        print("\n" + "="*50)
        print("SIMULADOR DE CACHE - MAPEAMENTO DIRETO")
        print("="*50)
        print("RAM: 32 endereços (00000 a 11111)")
        print("Cache: 4 linhas (Mapeamento: Tag(2) Linha(2) Byte(1))")

        addr = input("\nDigite o endereço binário de 5 bits (ou 'S' para sair): ").lower()
        if addr == 's':
            break

        resultado = sim.acessar(addr)
        print(resultado)

if __name__ == "__main__":
    menu()
```
