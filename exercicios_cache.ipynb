# **EXERCICIO 5.2**

```python
pra saber a quantidade de linhas precisamos pegar a capacidade da cache e dividir pelo numero de bytes por linha:
32KB = 32.768
32.768/8 = 4096 linhas ou 2^12(index de 12 bits)

bits por linha = 8 x 8 = 64bits

o tamanho do endereço fisico que o processador usa pra localizar um dado = 24 (16MB = 2^24)

pra apontar o byte exato = 3 (8 = 2^3)

a tag = 24 - 12 - 3
tag = 9

bit de validade = 1

total por linha = 64 + 9 + 1 = 74

bits nescessários pra memoria cache funcionar = 4096 x 74
= 303.104
```

# **EXERCICIO 5.3**

```python
para saber o quantidade de bits do endereço precisamos colocar a capacidade na base 2 e ver seu expoente.
capacidade = 64MB = 2^6 x 2^20 = 2^26
tamano do endereço = 26 bits

o tamano da palavra é o expoente quando colocamos a quantidade de bytes por linha na base 2
16B = 2^4
palavra = 4

o tamnaho da linha é igual ao expoente quando colocamos a quantidade de linhas na base 2
2K = 2^1 x 2^10
linha = 11

e a tag é a subtração do total com a palavra e a linha
então tag = 26 - 11 - 4
tag = 11

formato da palavra:
  TAG(11bits)  linha(11bits)  PALAVRA(4bits)
  00000000000   00000000000        0000
```

# EXERCICIO **5.6**

```python
o tamanho do endereço é dado pelo expoente quando a capacidade da MP é colocada na base 2
64MB  = 2^6 x 2^20
endereço de 26 bits

o expoente do numero da largura me diz o tamanho do campo da celula
16B = 2^4
celula de 4 bits

o total de blocos é dado quando botados a capacidade da MP sobre a largura:
2^26/2^4
blocos = 2^22 ou 4M
e o campo do bloco é formado pelo numero do expoente do total de blocos, ou seja, 22.

então o formato dessa memoria é:

     bloco (22bits)           palavra(4)
0000000000000000000000          0000
```
