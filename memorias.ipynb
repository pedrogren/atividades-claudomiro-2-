# **Memória 32x16 com 4 bancos**

```python
def simulador_chip():
    bancos = [
        ["0000000000000000"] * 8,
        ["0000000000000000"] * 8,
        ["0000000000000000"] * 8,
        ["0000000000000000"] * 8
    ]

    print("="*35)
    print("CHIP DE MEMÓRIA 32x16 (4 BANCOS)")
    print("="*35)
    print("Total: 32 endereços | Dado: 16 bits")
    print("Endereços: 0 a 31")

    while True:
        print("\nOpções: [W] Escrever | [R] Ler | [L] Listar | [S] Sair")
        opcao = input("Escolha uma opção: ").upper()

        if opcao == 'S':
            break

        if opcao in ['W', 'R']:
            try:
                ende = int(input("Digite o endereço (0-31): "))

                if ende < 0 or ende > 31:
                    print("Erro: Endereço fora do limite (0-31).")
                    continue
                num_banco = ende // 8
                ende_interno = ende % 8

                if opcao == 'W':
                    dado = input("Digite o dado (16 bits binários): ")
                    if len(dado) == 16:
                        bancos[num_banco][ende_interno] = dado
                        print(f"Sucesso! Gravado no Banco {num_banco}, Posição {ende_interno}")
                    else:
                        print("Erro: O dado deve ter exatamente 16 bits.")

                elif opcao == 'R':
                    dado_lido = bancos[num_banco][ende_interno]
                    print(f"Lendo Endereço {ende}:")
                    print(f"-> Localização: Banco {num_banco}, Endereço Interno {ende_interno}")
                    print(f"-> Conteúdo: {dado_lido}")

            except ValueError:
                print("Erro: Digite um número válido para o endereço.")

        elif opcao == 'L':
            for b in range(4):
                print(f"\nBANCO {b}:")
                for i in range(8):
                    ende_global = (b * 8) + i
                    print(f"  [{ende_global:02d}] {bancos[b][i]}")
        else:
            print("Opção inválida.")

if __name__ == "__main__":
    simulador_chip()
```
