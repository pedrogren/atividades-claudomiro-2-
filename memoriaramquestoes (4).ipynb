# **Questão 4.3**

Para resolver o **Exemplo 4.3**, precisamos analisar as características da Memória Principal (MP) fornecidas: uma capacidade de **2K palavras**, sendo que cada palavra possui **16 bits**. O problema pede o tamanho dos registradores **REM** e **RDM**, o valor do **maior endereço** e a **capacidade total** em bits.

## Tamanho dos Registradores REM e RDM

O **RDM (Registrador de Dados da Memória)** tem a função de armazenar temporariamente a informação que está sendo transferida entre o processador e a memória. O tamanho deste registrador deve ser igual ao tamanho da unidade de transferência da memória, que neste exemplo é a palavra de **16 bits**. Portanto, o **RDM possui 16 bits**.

Já o **REM (Registrador de Endereços da Memória)** é responsável por guardar o endereço da célula que será acessada para uma leitura ou escrita. O seu tamanho em bits depende da quantidade total de endereços disponíveis na memória. Como a memória tem **2K endereços** e sabemos que **1K** equivale a **1024** ($2^{10}$), o total de endereços é **2048** ($2^{11}$). Seguindo a regra de que o número de endereços $N = 2^E$ (onde $E$ é o número de bits), descobrimos que o **REM possui 11 bits**.

## Valor do Maior Endereço da MP

Em qualquer memória organizada sequencialmente, os endereços começam no número **0** e vão até o total de células menos um, ou seja, **N - 1**.

Como calculamos que esta memória possui um total de **2048 posições** de armazenamento (2K), basta subtrair um para encontrar o último endereço:
*   **2048 - 1 = 2047**.

Portanto, o **valor do maior endereço** dessa memória é **2047** em decimal. Se representarmos esse valor em binário, ele será formado por uma sequência de **11 bits** (todos com valor 1), que é exatamente a capacidade total do registrador **REM**.

## Quantidade Total de Bits Armazenados

A quantidade total de bits (**T**) que uma memória pode guardar é o resultado da multiplicação do número de células (**N**) pelo tamanho de cada célula (**M**).

Fazendo a conta para este exemplo:
*   **N** (total de células) = **2048** (ou 2K).
*   **M** (tamanho da célula) = **16 bits**.
*   **T = 2048 x 16 = 32.768 bits**.

Para facilitar a leitura, podemos expressar esse valor na unidade **K** (quilo). Como **1K** bits são **1024** bits, dividimos o total por 1024:
*   **32.768 / 1024 = 32**.

A **quantidade total de bits** que podem ser armazenados nesta memória é de **32K bits**.

# **Questão 5**

## Diferenças entre as Memórias A, B e C

Para entender a diferença entre essas três organizações de memória, precisamos olhar para três coisas: o tamanho do **endereço** (quantos bits o processador usa para achar a célula), o **conteúdo** (quantos bits cabem em cada "caixinha") e o **total de bits** (a capacidade total de armazenamento).

## Memória A: 32K células de 8 bits

Nessa organização, temos **32.768 células** (32 x 1024). Como 32K é o mesmo que **2 elevado a 15**, o tamanho do **endereço** é de **15 bits**. O **conteúdo** de cada célula é de **8 bits** (ou 1 byte). Multiplicando as células pelo tamanho de cada uma, o **total de bits** é de **256K bits** (32 x 8).

## Memória B: 16K células de 16 bits

Aqui temos metade das células da Memória A, ou seja, **16.384 células** (16 x 1024). Como 16K é **2 elevado a 14**, o **endereço** é menor, com **14 bits**. Por outro lado, o **conteúdo** de cada célula é o dobro, com **16 bits**. No fim das contas, o **total de bits** é o mesmo da Memória A: **256K bits** (16 x 16), pois o que ela perdeu em número de endereços, ganhou no tamanho da célula.

## Memória C: 16K células de 8 bits

Esta é a menor das três. Ela tem as mesmas **16.384 células** da Memória B, o que significa que o seu **endereço** também tem **14 bits**. No entanto, o **conteúdo** de cada célula é de apenas **8 bits**. Por isso, o seu **total de bits** é de apenas **128K bits** (16 x 8), que é a metade da capacidade das memórias A e B.

## Resumo das diferenças

Comparando as três, a **Memória A** é a que precisa de mais bits para o **endereço** (15 bits). A **Memória B** é a que tem mais **conteúdo** por célula (16 bits). Já o **total de bits** mostra que as memórias **A e B** empatam na capacidade total, enquanto a **Memória C** é a mais limitada.

# **Questão 6**

## Função do Registrador de Endereços de Memória (REM)

O **REM** tem a função de guardar temporariamente o **endereço** da célula de memória que o processador deseja acessar. Quando o computador vai iniciar uma operação de leitura ou de escrita, ele coloca nesse registrador o número da "casa" (posição) onde a informação está ou onde ela deve ser guardada. Ele serve como uma ponte que indica para a memória qual é o local exato da operação.

## Função do Registrador de Dados da Memória (RDM)

O **RDM** serve para armazenar temporariamente a **informação** propriamente dita que está viajando entre a memória principal e o processador. Se o processador está lendo algo da memória, o dado sai da célula e fica no RDM até o processador pegá-lo. Se o processador está gravando algo, ele coloca o dado no RDM primeiro para depois enviá-lo para a memória. Por isso, ele é chamado de **bidirecional**, já que os dados podem passar por ele tanto para entrar quanto para sair da memória.

# **Questão 7**

## Os três caminhos de comunicação entre o processador e a memória

A comunicação entre o cérebro do computador (processador) e a sua memória principal acontece através de três tipos de barramentos, que funcionam como "estradas" para os dados e sinais. Cada um desses caminhos tem uma função muito específica e uma regra sobre para qual lado a informação pode viajar.

## O Barramento de Dados

O **barramento de dados** é o caminho responsável por carregar a informação propriamente dita, como um número ou uma letra, entre a Memória Principal e o processador. Ele faz a ligação entre o componente chamado RDM (Registrador de Dados da Memória) e a memória para que as transferências ocorram. Como o computador precisa tanto buscar um dado guardado quanto salvar um dado novo, esse barramento é **bidirecional**, o que significa que o fluxo de sinais pode ir e voltar pelos mesmos fios.

## O Barramento de Endereços

O **barramento de endereços** tem a função exclusiva de levar os bits que representam o endereço da célula que o processador quer acessar. Ele conecta o registrador REM (Registrador de Endereços da Memória) diretamente à Memória Principal. Diferente do anterior, o fluxo de sinais aqui é **unidirecional**, pois apenas o processador indica para a memória onde ele quer mexer; a memória não envia endereços de volta para o processador.

## O Barramento de Controle

O **barramento de controle** é formado por fios condutores que servem para organizar e comandar o que deve ser feito no sistema. Por ele passam sinais fundamentais como o de escrita (**WRITE**), o de leitura (**READ**). Esse conjunto de fios interliga o processador à memória para coordenar essas tarefas e o seu fluxo de sinais é **bidirecional**, permitindo que as ordens e as respostas de estado circulem entre os dois lados.

Para resolver essa questão,precisamos comparar o **custo total** dos dois sistemas de memória. A ideia é descobrir a partir de que ponto a economia no preço por bit da DRAM compensa o custo mais alto da interface dela.

Aqui estão os dados que o problema passou:\
Interface DRAM: R 5,00\
Interface SRAM: R 1,00\
Custo por bit DRAM: R 0,00001\
Custo por bit SRAM: R 0,00002

### 1. Montando as equações de custo
Vamos chamar a quantidade de bits de $x$. O custo total de cada sistema é a soma do preço da interface com o preço total dos bits:

*   **Custo SRAM** $= 1,00 + 0,00002 \cdot x$
*   **Custo DRAM** $= 5,00 + 0,00001 \cdot x$

### 2. Comparando os custos
Para que a memória DRAM seja mais barata, o custo total dela precisa ser menor que o da SRAM:
$$5,00 + 0,00001x < 1,00 + 0,00002x$$

### 3. Isolando o $x$
Agora é só fazer a álgebra básica:
$$5,00 - 1,00 < 0,00002x - 0,00001x$$
$$4,00 < 0,00001x$$

Dividindo 4 por 0,00001:
$$x > \frac{4}{0,00001}$$
$$x > 400.000$$

### Conclusão
O conjunto de memória dinâmica (DRAM) passa a ser mais barato quando o sistema tem **mais de 400.000 bits**.

Para resolver essa questão, primeiro anotamos os valores que o enunciado nos passou:
* **Tempo de acesso à Cache ($T_{cache}$):** 8 ns.
* **Tempo de acesso à Memória Principal ($T_{main}$):** 70 ns.

* **Número de acessos analisados:** 100.

## Entendendo a lógica do acesso
O processador sempre tenta buscar a informação primeiro na **memória cache**, que é o nível mais rápido e caro da hierarquia. Se o dado estiver lá (acerto ou *hit*), o acesso é instantâneo. Se não estiver (falta ou *miss*), o sistema perde o tempo da busca na cache e ainda precisa gastar o tempo de busca na **memória principal**.

Nesse sistema específico, temos dois caminhos possíveis:
1. **Cenário de Acerto (Hit):** Acontece em 96 dos 100 acessos. O tempo gasto é apenas de **8 ns**.
2. **Cenário de Falta (Miss):** Acontece nos outros 4 acessos (100 - 96). O tempo gasto é o tempo de olhar a cache (**8 ns**) mais o tempo de buscar na memória principal (**70 ns**), totalizando **78 ns**.

## Cálculo passo a passo do tempo médio
Vamos calcular o tempo total que o processador gastou para realizar todos os 100 acessos e depois tirar a média simples:

**1. Cálculo do tempo total dos acertos:**
* 96 acessos $\times$ 8 ns = **768 ns**.

**2. Cálculo do tempo total das faltas:**
* 4 acessos $\times$ 78 ns = **312 ns**.

**3. Soma do tempo total acumulado:**
* 768 ns (acertos) + 312 ns (faltas) = **1080 ns**.

**4. Divisão pelo número de acessos para achar a média:**
* $1080 \text{ ns} / 100 \text{ acessos} = \mathbf{10,8 \text{ ns}}$.

A resposta final é que o **tempo médio de acesso do sistema é de 10,8 ns**. Esse resultado demonstra como uma cache eficiente consegue manter a velocidade do sistema muito próxima da velocidade da memória mais rápida, mesmo quando a memória principal é quase dez vezes mais lenta.

# **Questão 33**

## Poder de processamento vs Capacidade de armazenamento

Essa afirmação **não é válida** porque ela mistura dois conceitos de hardware que funcionam de formas independentes dentro de um computador. O **poder de processamento** está ligado diretamente à Unidade Central de Processamento (UCP ou CPU) e à sua velocidade para executar cálculos e instruções por segundo. Já o ato de **armazenar programas** é uma função das unidades de memória, como o disco rígido (HD), o SSD ou a própria memória principal.

## Por que a velocidade não define o espaço

Um computador pode ter um processador extremamente rápido e moderno, mas se ele tiver um disco rígido com pouca **capacidade**, ele não conseguirá guardar muitos programas. Da mesma forma, um computador antigo e lento pode ter um HD imenso e conseguir armazenar milhares de arquivos e softwares, mesmo que demore muito tempo para conseguir abrir cada um deles. Na **hierarquia de memória**, o processador fica no topo buscando velocidade, enquanto a base da pirâmide (memória secundária) foca em oferecer o máximo de espaço possível pelo menor custo.

## A única ligação técnica entre os dois

Existe apenas um detalhe técnico onde o processador limita o armazenamento, mas isso tem a ver com a sua **arquitetura** e não com o seu "poder" ou velocidade. O tamanho do **endereçamento** do processador define o máximo de memória RAM (memória principal) que o sistema consegue identificar. Por exemplo, um processador de **32 bits** só consegue "enxergar" e usar até **4GB de RAM**, enquanto um de **64 bits** consegue gerenciar uma quantidade absurdamente maior, chegando a **16 Exabytes**. Mesmo assim, isso define o limite de quanto o computador pode usar ao mesmo tempo, e não a quantidade total de programas que você pode deixar instalados no seu disco rígido.

# **Questão 34**

## Validade da afirmação sobre a Memória Principal

Essa afirmação **não é tecnicamente válida**. Aumentar a capacidade da Memória Principal (RAM) não torna o acesso aos meios magnéticos, como o disco rígido, fisicamente mais rápido. O **tempo de acesso** de um dispositivo é definido pela sua tecnologia de construção e pela velocidade de seus próprios circuitos. Se um disco rígido (HD) leva entre **8 e 30 milissegundos** para localizar uma informação, ele continuará levando esse mesmo tempo, independentemente de haver mais ou menos memória RAM disponível no sistema.

## Por que o computador parece mais veloz

O que ocorre na prática é uma melhora no **desempenho geral** do sistema, e não um aumento na velocidade do disco em si. De acordo com a **hierarquia de memória**, a Memória Principal é muito mais rápida (acesso entre **50 e 80 nanossegundos**) do que a memória secundária (HDs e disquetes). Quando você aumenta a RAM, o computador consegue manter mais dados "carregados" nela, evitando que o processador precise buscar informações com frequência na base da pirâmide, que é a parte mais lenta do sistema.

## Diferença entre as tecnologias

É importante lembrar que essas memórias utilizam tecnologias totalmente distintas: a RAM é uma **memória de semicondutores**, enquanto os discos rígidos e disquetes são de **meio magnético**. Como o acesso aos meios magnéticos envolve componentes mecânicos que são muito mais lentos, a única forma de realmente acelerar o acesso a esses dados seria trocar o dispositivo por uma tecnologia superior, como um **SSD**, que possui velocidades de transferência significativamente maiores que os HDs tradicionais.

# **Questão 26**

### 1. Estrutura de Endereçamento
Para endereçar 8 células, o sistema utiliza **3 bits de endereço** ($A_2, A_1, A_0$). Esses bits entram em um **decodificador de 3 para 8**.
*   **Função do Decodificador:** Ele possui 8 saídas ($D_0$ a $D_7$). Quando um endereço binário entra, apenas a saída correspondente àquele número é ativada (vai para nível 1).
*   **Pino de Habilitação (Enable):** Seguindo o modelo da figura, existe um pino que deve estar em **1** para que qualquer saída do decodificador seja ativada.

### 2. Mapeamento da Memória (Tabela de Conteúdo)
Este é o conteúdo "gravado" fisicamente na sua simulação:

| Endereço ($A_2 A_1 A_0$) | Linha Ativa | Dado Armazenado ($S_3 S_2 S_1 S_0$) |
| :--- | :--- | :--- |
| 000 | $D_0$ | **1010** |
| 001 | $D_1$ | **1100** |
| 010 | $D_2$ | **0110** |
| 011 | $D_3$ | **0011** |
| 100 | $D_4$ | **1111** |
| 101 | $D_5$ | **0001** |
| 110 | $D_6$ | **1001** |
| 111 | $D_7$ | **0101** |

### 3. Matriz de Portas OU (Lógica de Saída)
As expressões lógicas que você usou para interligar as saídas do decodificador às portas OU de saída são:

*   **Saída $S_3$:** $D_0 + D_1 + D_4 + D_6$ (Bit mais significativo)
*   **Saída $S_2$:** $D_1 + D_2 + D_4 + D_7$
*   **Saída $S_1$:** $D_0 + D_2 + D_3 + D_4$
*   **Saída $S_0$:** $D_3 + D_4 + D_5 + D_6 + D_7$ (Bit menos significativo)

### 4. Funcionamento do Circuito
Diferente de uma RAM, essa ROM não possui pinos de escrita (como o $R/W$ ou $WE$). O processo é puramente de **leitura**:
1.  Você coloca o endereço nos pinos de entrada.
2.  O decodificador "aponta" para a linha da célula desejada.
3.  As conexões físicas (representadas pelas expressões acima) conduzem o sinal de nível 1 até as portas OU correspondentes.
4.  O dado de 4 bits aparece instantaneamente nas saídas $S_3, S_2, S_1$ e $S_0$.
