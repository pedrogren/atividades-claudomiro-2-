# **4096 k x 1**

```python
TOTAL_ENDERECOS = 2**22
TOTAL_LINHAS = 2**11
TOTAL_COLUNAS = 2**11
LARGURA_BITS = 1
BITS_ENDERECOS = 22

memoria = [["0"] * TOTAL_COLUNAS for _ in range(TOTAL_LINHAS)]

def ler_endereco():
    return int(input(f"Endereco (0 a {TOTAL_ENDERECOS - 1}): "))

def ler_dado():
    return input(f"Dado ({LARGURA_BITS} bit): ")

def decompor(endereco):
    linha = endereco // TOTAL_COLUNAS
    coluna = endereco % TOTAL_COLUNAS
    return linha, coluna

def escrever(endereco, dado):
    linha, coluna = decompor(endereco)
    memoria[linha][coluna] = dado
    print(f"Gravado em linha {linha}, coluna {coluna}.")

def ler(endereco):
    linha, coluna = decompor(endereco)
    valor = memoria[linha][coluna]
    print(f"Linha {linha}, coluna {coluna}: {valor}")


print("MEMORIA 4096K x 1")

comandos = {"W": "escrever", "R": "ler", "Q": "sair"}
ativo = True

while ativo:
    acao = input("Digite W (escrever), R (ler) ou Q (sair): ").strip().upper()

    if acao == "W":
        end = ler_endereco()
        dado = ler_dado()
        escrever(end, dado)

    elif acao == "R":
        end = ler_endereco()
        ler(end)

    elif acao == "Q":
        ativo = False

    else:
        print("Opcao invalida!")
```

# **128M X 4**

```python
TOTAL_ENDERECOS = 2**27
TOTAL_LINHAS = 2**13
TOTAL_COLUNAS = 2**13
LARGURA_BITS = 4
N_BANCOS = 2

BITS_ENDERECOS = 27
BITS_BANCO = 1

memoria = [
    [[["0"] * LARGURA_BITS for _ in range(TOTAL_COLUNAS)] for _ in range(TOTAL_LINHAS)]
    for _ in range(N_BANCOS)
]

def ler_endereco():
    return int(input(f"Endereco (0 a {TOTAL_ENDERECOS - 1}): "))

def ler_dado():
    return input(f"Dado ({LARGURA_BITS} bits): ")

def decompor(endereco):
    banco = endereco // (2**26)
    resto = endereco % (2**26)
    linha = resto // TOTAL_COLUNAS
    coluna = resto % TOTAL_COLUNAS
    return banco, linha, coluna

def escrever(endereco, dado):
    banco, linha, coluna = decompor(endereco)
    memoria[banco][linha][coluna] = dado
    print(f"Gravado no banco {banco}, linha {linha}, coluna {coluna}.")

def ler(endereco):
    banco, linha, coluna = decompor(endereco)
    valor = memoria[banco][linha][coluna]
    print(f"Banco {banco}, linha {linha}, coluna {coluna}: {valor}")


print("MEMORIA COM 2 BANCOS")

ativo = True

while ativo:
    acao = input("Digite W (escrever), R (ler) ou Q (sair): ").strip().upper()

    if acao == "W":
        end = ler_endereco()
        dado = ler_dado()
        escrever(end, dado)

    elif acao == "R":
        end = ler_endereco()
        ler(end)

    elif acao == "Q":
        ativo = False

    else:
        print("Opcao invalida!")
```
