```python
def simulador_memoria():
    enderecos = [format(i, '04b') for i in range(16)]
    memoria = {ende: "00000000" for ende in enderecos}

    while True:
        print("\n" + "="*65)
        print("SIMULADOR DE MEMÓRIA RAM - ORGANIZAÇÃO E ARQ. DE COMPUTADORES")
        print("="*65)
        opcao = input("Digite W para escrever, R para ler, L para listar toda a memoria e qualquer tecla para parar: ").upper()

        if opcao == 'W':
            ende = input("Digite o endereço (4 bits, ex: 0000): ")
            if ende in memoria:
                dado = input("Digite o dado (8 bits, ex: 10101010): ")
                if len(dado) == 8 and all(c in '01' for c in dado):
                    memoria[ende] = dado
                    print(f"Gravado com sucesso no endereço {ende}!")
                else:
                    print("Erro: O dado deve ter 8 bits binários.")
            else:
                print("Erro: Endereço inválido.")

        elif opcao == 'R':
            ende = input("Digite o endereço para leitura (4 bits): ")
            if ende in memoria:
                print(f"Conteúdo no endereço {ende}: {memoria[ende]}")
                if memoria[ende] == "00000000":
                    print("Esse Endereço está totalmente zerado, ou seja não foi gravado nada nele!")
            else:
                print("Erro: Endereço inválido.")

        elif opcao == 'L':
            print("\nTodos os estados:")
            for ende in enderecos:
                print(f"{ende}: {memoria[ende]}")

        else:
            print("Encerrando o simulador...")
            break

if __name__ == "__main__":
    simulador_memoria()
```
